````python
df = fdr.DataReader('088350')
df.head()
````

````python
df=df.rename(columns={'Open': '시가','High': '고가','Low':'저가','Close': '종가','Volume': '거래량','Change': '변동률'})
df.head()
````

````python
#수익률=(매도가격-매수가격)/매수가격
df['수익률']= df['종가'].pct_change()
df['10Days']=df['종가'].rolling(window=10).mean()
df['50Days']=df['종가'].rolling(window=50).mean()
df.head(50)

````

````python
df=df.dropna() #nan 지우기
df.head()
````

````python
import numpy as np #숫자계산을 위한 도구(np)
df['target'] = np.where(df['수익률'].shift(-1)<0,1,0)
df.head() #공매도가 일어나려면 수익률이 음수여야 하는데 공매도가 될때 데이터를 사용하려면 하나씩 당겨서 와야하기 떄문
````

````python

#데이터 분할
from sklearn.model_selection import train_test_split #import sklearn을 쓰기에는 너무 커서 지정
X_train,X_test,y_train,y_test=train_test_split(
    df[['수익률','10Days','50Days']],df['target'],test_size=0.2,random_state=42) #random_state: 데이터 치우침을 막기위해 데이터를 섞는다.대부분 42사용
````

````python
print(y_train.shape,X_train.shape)
````

````python
from sklearn.linear_model import LogisticRegression
lr =LogisticRegression() #LogisticRegression :어떤 값을 넣던 0 또는 1로 바꾼다
````

````python
lr.fit(X_train,y_train)    #fit 학습시키는 함수 이함수로 인공지능 생성
````

````python
lr.predict(X_test)
````
